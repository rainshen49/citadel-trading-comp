```python
import json
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open("./prices2.json") as datafile:
    prices = json.load(datafile)
with open("./imbalances2.json") as datafile:
    imbalances = json.load(datafile)
```

```python
deltas = {ticker: [0] + [prices[ticker][i]-prices[ticker][i-1] for i in range(1,len(prices[ticker]))] for ticker in prices}
```

```python
for ticker, delta in deltas.items():
    deltamax = max(delta)
    delta = [d/deltamax for d in delta]
    imbalancesmax = max(imbalances[ticker])
    imbalance = [d/imbalancesmax for d in imbalances[ticker]]
    f = plt.figure(figsize=(20,20))
    plt.plot(imbalance)
    plt.plot(delta)
```
